```python
import pandas as pd
from ftlangdetect import detect
from ast import literal_eval
```

```python
df = pd.read_csv("/Users/FrancesY/Documents/Columbia/fall2022/5067nlp/data/DatingAppReviewsDataset.csv", index_col=0)
```

```python
def det(x):
    try:
        lang = detect(x,low_memory=True)
    except:
        lang = 'Other'
    return lang

df['Lang'] = df['Review'].apply(det)
```

```python
df2 = pd.io.json.json_normalize(df.Lang)
```

```python
df = df.reset_index().drop("index", axis = 1)
```
