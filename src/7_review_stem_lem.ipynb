## 7. Text Cleaning, Removing stopwords, Stemming and lemmatization

```python
import pandas as pd
import numpy as np
import re
import nltk
from nltk.corpus import stopwords 
from sklearn.feature_extraction.text import CountVectorizer
from nltk.stem import WordNetLemmatizer
from utils import *
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('df_w_sentiments_nov17_franyang.csv')
```

```python
df.groupby('App')['Rating'].describe()
```

```python
# Clean text
df['review_clean'] = df['Review'].apply(clean_text)
```

```python
# Remove stopwords
df['review_clean'] = df['review_clean'].apply(rem_sw)
```

```python
# Stemming
df['review_stem'] = df['review_clean'].apply(my_stem)
```

```python
# Lemmatization
wordnet_lem = WordNetLemmatizer()
df['review_lem'] = df['review_clean'].apply(wordnet_lem.lemmatize)
```

```python
# Stemming and lemmatization
df['review_stem_lem'] = df['review_stem'].apply(wordnet_lem.lemmatize)
```
